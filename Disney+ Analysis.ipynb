# Preprocessing the data


```python
df['date_added'] = pd.to_datetime(df['date_added'])  # converting the date to proper date format
```

```python
df['year_added'] = df['date_added'].dt.year # extracting year in separate column
df['year_added'][0:5]
```

```python
df['month_added'] = df['date_added'].dt.month # extracting month in separate column
df['month_added'][0:5]
```

```python
df['duration'] = df['duration'].str.extract(r'(\d+)').astype(float)  # converting the duration to float datatype
df['duration']
```

```python
plt.figure(figsize = (6,4))
sns.countplot(data = df,x = 'type')
plt.title("Distribution of Movies and TV Shows")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show() # no of movies vs tv shows
```

```python
plt.figure(figsize = (10,5))
sns.histplot(df['release_year'],bins = 30,kde = True,color = 'skyblue')
plt.title("Movies and TV shows Release Trend")
plt.xlabel("Release year")
plt.ylabel("Count")
plt.show()
```

```python
from collections import Counter
```

```python
all_genres = [genre.strip() for sublist in df['listed_in'] for genre in sublist.split(',')]
genre_counts = pd.Series(Counter(all_genres)).sort_values(ascending = False)

plt.figure(figsize = (10,5))
sns.barplot(x = genre_counts.index[:10], y = genre_counts.values[:10],hue = genre_counts.index[:10],palette = 'viridis')
plt.title("Top 10 most common genres")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.figure(figsize = (8,4))
sns.countplot(data = df,y = 'rating',hue = 'rating',palette = 'coolwarm',order = df['rating'].value_counts().index)
plt.title("Distribution of Ratings")
plt.xlabel("Count")
plt.ylabel("Rating")
plt.show()
```

```python
top_countries = df['country'].value_counts().nlargest(10)

plt.figure(figsize = (10,5))
sns.barplot(x = top_countries.index, y = top_countries.values, hue = top_countries.index ,palette = 'muted')
plt.title("Top 10 Countries with Most Disney+ Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

```python
top_directors = df['director'].value_counts().nlargest(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_directors.values, y=top_directors.index, hue = top_directors.index,palette="rocket")
plt.title("Top 10 Most Frequent Directors")
plt.xlabel("Count")
plt.ylabel("Director")
plt.show()
```

```python
all_actors = [actor.strip() for sublist in df['cast'].str.split(',') for actor in sublist]
actor_counts = pd.Series(Counter(all_actors)).sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=actor_counts.values, y=actor_counts.index,hue = actor_counts.values, palette="magma")
plt.title("Top 10 Most Featured Actors")
plt.xlabel("Count")
plt.ylabel("Actor")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['year_added'], bins=15, kde=True, color="green")
plt.title("Disney+ Content Addition Over Years")
plt.xlabel("Year Added")
plt.ylabel("Number of Titles")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x='release_year', y='duration', alpha=0.5)
plt.title("Movie Duration Over Time")
plt.xlabel("Release Year")
plt.ylabel("Duration (min)")
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['listed_in'].fillna(' '))

kmeans = KMeans(n_clusters = 5,random_state = 42,n_init = 10)
df['cluster'] = kmeans.fit_predict(X)


sns.countplot(x = df['cluster'],hue = df['cluster'],palette = "tab10")
plt.title("Cluster Distribution of Movies by Genre")
plt.show()
```
